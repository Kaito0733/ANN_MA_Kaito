```python
import numpy as np

def lrp(weights, activations, R):
    relevances = [None] * (len(weights) + 1)
    relevances[-1] = R
    
    for l in reversed(range(len(weights))):
        W = weights[l]
        A = activations[l]
        Z = W * A
        Zs = Z.sum(axis=1, keepdims=True) 
        S = relevances[l+1] / Zs   
        C = (Z * S).sum(axis=0)    
        relevances[l] = C          
    return relevances
```
